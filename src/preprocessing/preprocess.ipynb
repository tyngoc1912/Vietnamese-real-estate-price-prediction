## d. Tạo thuộc tính ngày đăng gốc

**Giải thích:**

  + Thuộc tính *post date* : bài đăng chỉ cung cấp nội dung về thời gian đăng như "Đăng hôm nay", "Đăng hôm qua", "Đăng x ngày trước",... nên quá trình thống kê thời gian đăng bài gặp khó khăn
  + Thuộc tính *metadata.Date* : lưu ngày tháng năm thu thập dữ liệu

Như vậy thông qua hai thuộc tính này, nhóm đã tiến hành tạo ra thuộc tính ngày tháng năm đăng bài viết chính xác nhất

    --> **original_post_date**

```python
def compute_original_date(post_text, meta_date):
    """
    Xác định ngày đăng bài gốc (original post date) từ nội dung văn bản bài đăng
    và ngày metadata đi kèm.

    Hàm này trích xuất thông tin ngày đăng từ chuỗi `post_text` (thường là mô tả
    thời gian như "Đăng hôm nay", "Đăng 3 ngày trước", hoặc ngày cụ thể),
    sau đó quy đổi về một đối tượng datetime dựa trên `meta_date`.

    Các trường hợp được hỗ trợ:
    1. "Đăng hôm nay"
       → Trả về chính `meta_date`

    2. "Đăng hôm qua" / "Đăng hôm q"
       → Trả về `meta_date - 1 ngày`

    3. "Đăng x ngày trước" (x là số nguyên)
       → Trả về `meta_date - x ngày`

    4. Ngày dạng "dd/mm/yyyy"
       → Trả về datetime(y, m, d)

    5. Ngày dạng ISO "yyyy-mm-dd"
       → Trả về datetime(y, m, d)

    Thứ tự ưu tiên:
    - Các biểu thức thời gian tương đối (hôm nay, hôm qua, x ngày trước)
    - Sau đó mới đến ngày tuyệt đối (dd/mm/yyyy, yyyy-mm-dd)

    Parameters
    ----------
    post_text : str
        Chuỗi văn bản chứa thông tin ngày đăng bài
        (ví dụ: "Đăng 2 ngày trước", "Đăng hôm nay", "01/03/2023").

    meta_date : datetime hoặc str
        Ngày metadata của bài đăng (ví dụ: ngày crawl dữ liệu).
        Nếu là chuỗi, sẽ được chuyển sang datetime bằng `pd.to_datetime`.

    Returns
    -------
    datetime hoặc None
        - Trả về đối tượng `datetime` biểu diễn ngày đăng bài gốc nếu trích xuất thành công.
        - Trả về None nếu:
          + `meta_date` không hợp lệ
          + Không khớp bất kỳ mẫu ngày nào trong `post_text`.

    Notes
    -----
    - Hàm không xử lý múi giờ (timezone-naive datetime).
    - Văn bản `post_text` được chuyển về chữ thường (lowercase) trước khi xử lý.
    - Phù hợp để dùng trong tiền xử lý dữ liệu (feature engineering) cho bài toán NLP,
      crawling hoặc phân tích dữ liệu bất động sản / mạng xã hội.
    """
    # meta_date phải là datetime
    if isinstance(meta_date, str):
        meta_date = pd.to_datetime(meta_date, errors='coerce')
    if pd.isna(meta_date):
        return None

    text = str(post_text).strip().lower()
    # ---- "Đăng hôm nay" ----
    if "đăng hôm nay" in text:
        return meta_date

    # ---- "Đăng hôm qua" ----
    if "đăng hôm qua" in text or "đăng hôm q" in text:
        return meta_date - timedelta(days=1)

    # ---- "Đăng x ngày trước" ----
    match_days = re.search(r'đăng\s+(\d+)\s+ngày', text)
    if match_days:
        days = int(match_days.group(1))
        return meta_date - timedelta(days=days)

    # ---- "dd/mm/yyyy" ----
    match_ddmmyyyy = re.search(r'(\d{1,2})/(\d{1,2})/(\d{4})', text)
    if match_ddmmyyyy:
        d, m, y = match_ddmmyyyy.groups()
        return datetime(int(y), int(m), int(d))

    # ---- "yyyy-mm-dd" ----
    match_iso = re.search(r'(\d{4})-(\d{2})-(\d{2})', text)
    if match_iso:
        y, m, d = match_iso.groups()
        return datetime(int(y), int(m), int(d))

    return None
```

## e. Chuẩn hoá thuộc tính "price_per_m2"

```python
def convert_price_million(text):
    """
    Chuẩn hóa giá tiền về đơn vị triệu (million) từ chuỗi văn bản mô tả giá.

    Hàm này trích xuất giá trị số đầu tiên xuất hiện trong chuỗi `text`,
    sau đó quy đổi về đơn vị **triệu** dựa trên đơn vị được đề cập trong văn bản.

    Các trường hợp được hỗ trợ:
    1. Đơn vị "ngàn" / "ngan" (ví dụ: "850 ngàn/m2")
       → Giá trị được chia cho 1000 để chuyển sang triệu
       (850 ngàn = 0.85 triệu)

    2. Đơn vị "triệu" / "trieu" (ví dụ: "12.5 triệu/m2")
       → Giữ nguyên giá trị số

    3. Không xác định rõ đơn vị
       → Trả về giá trị số thô đã trích xuất (giả định đã ở đơn vị triệu)

    Parameters
    ----------
    text : str
        Chuỗi văn bản mô tả giá tiền
        (ví dụ: "12 triệu/m2", "850 ngàn/m2", "5.5 trieu").

    Returns
    -------
    float hoặc None
        - Giá trị giá tiền đã được chuẩn hóa về đơn vị triệu.
        - Trả về None nếu không tìm thấy giá trị số trong chuỗi.

    Notes
    -----
    - Hàm chỉ trích xuất **số đầu tiên** xuất hiện trong chuỗi.
    - Văn bản được chuyển sang chữ thường (lowercase) trước khi xử lý.
    - Hàm không phân biệt đơn vị theo diện tích (ví dụ: /m2), chỉ xử lý giá trị tiền.
    - Phù hợp dùng trong bước tiền xử lý dữ liệu (feature engineering)
      cho bài toán phân tích giá bất động sản.
    """
    t = str(text).lower()

    # Tách số
    match = re.search(r'\d+\.?\d*', t)
    if not match:
        return None

    num = float(match.group())

    # Nếu đơn vị là "ngàn/m2" → chia 1000 để chuyển sang triệu
    if "ngàn" in t or "ngan" in t:
        return num / 1000

    # Nếu đơn vị là "triệu/m2" → giữ nguyên
    if "triệu" in t or "trieu" in t:
        return num

    # Nếu không xác định rõ đơn vị → trả số thô
    return num
```

```python
df['price_per_m2_million'] = df['price_per_m2'].apply(convert_price_million)
df
```

Sau khi thực hiện hàm **convert_price_million**, nhóm đã tạo được thuộc tính *price_per_m2_million* từ thuộc tính *price_per_m2*

## f. Chuẩn hoá thuộc tính "Mức giá"

```python
def convert_price(text):
    """
    Chuẩn hóa giá bất động sản về đơn vị **tỷ đồng** từ chuỗi văn bản mô tả giá.

    Hàm này trích xuất giá trị số đầu tiên xuất hiện trong chuỗi `text`,
    sau đó quy đổi về **đơn vị tỷ** dựa trên đơn vị tiền tệ được đề cập.

    Quy ước quy đổi:
    - "triệu"  → chia cho 1000 để chuyển sang tỷ
      (ví dụ: 500 triệu = 0.5 tỷ)
    - "tỷ" / "ty" → giữ nguyên giá trị
    - Không xác định được đơn vị → trả về None

    Parameters
    ----------
    text : str
        Chuỗi văn bản mô tả giá
        (ví dụ: "1.2 tỷ", "750 triệu", "3 ty").

    Returns
    -------
    float hoặc None
        - Giá trị giá đã được chuẩn hóa về **tỷ đồng**.
        - Trả về None nếu:
          + Không tìm thấy số trong chuỗi
          + Không xác định được đơn vị tiền tệ phù hợp.

    Notes
    -----
    - Hàm chỉ trích xuất **số đầu tiên** xuất hiện trong chuỗi.
    - Văn bản được xử lý không phân biệt chữ hoa / chữ thường.
    - Hàm không xử lý các trường hợp phức tạp như:
        + "1 tỷ 200 triệu"
        + "khoảng 2–3 tỷ"
        + "giá thỏa thuận"
    - Phù hợp dùng trong bước tiền xử lý dữ liệu (feature engineering)
      cho các bài toán phân tích giá bất động sản.
    """

    # Tìm số
    match = re.search(r'\d+\.?\d*', str(text))
    if not match:
        return None   # không có số → trả về None

    num = float(match.group())

    # Quy đổi
    t = str(text).lower()

    if "triệu" in t:
        return num / 1000

    if "tỷ" in t or "ty" in t:
        return num

    return None
```

```python
# Tạo cột mới: giá trị quy đổi về Tỷ
df["price"] = df["details.Mức giá"].apply(convert_price)
df
```

## g. Chuẩn hoá thuộc tính "Diện tích"

```python
def convert_area(text):
    """
    Trích xuất và chuẩn hóa diện tích từ chuỗi văn bản.

    Hàm này tìm **giá trị số đầu tiên** xuất hiện trong chuỗi `text`
    và trả về dưới dạng số thực (float), thường dùng để biểu diễn
    diện tích (m²).

    Parameters
    ----------
    text : str
        Chuỗi văn bản mô tả diện tích
        (ví dụ: "82m2", "diện tích 45.5 m²", "khoảng 100 mét vuông").

    Returns
    -------
    float hoặc None
        - Giá trị diện tích đã trích xuất (float).
        - Trả về None nếu không tìm thấy giá trị số trong chuỗi.

    Notes
    -----
    - Hàm không kiểm tra đơn vị diện tích (m2, m², mét vuông...).
    - Chỉ lấy **số đầu tiên** xuất hiện trong chuỗi.
    - Không xử lý các trường hợp phức tạp như:
        + "80–100m2"
        + "2 x 5 = 10m2"
    - Phù hợp dùng trong bước tiền xử lý dữ liệu (feature engineering)
      cho các bài toán phân tích bất động sản.
    """
    match = re.search(r'\d+\.?\d*', str(text))
    return float(match.group()) if match else None
```

```python
df['area'] = df['details.Diện tích'].apply(convert_area)
df
```

```python
def classify_legal(text):
    """
    Phân loại tình trạng pháp lý bất động sản từ mô tả văn bản.

    Hàm này dựa trên việc dò tìm các từ khóa pháp lý phổ biến trong chuỗi `text`
    để gán nhãn tình trạng pháp lý của bất động sản theo các nhóm chuẩn hóa.

    Các từ khóa được nhận diện:
    - "sổ hồng", "sổ đỏ"
    - "hợp đồng mua bán"
    - "giấy viết tay"
    - "đang chờ sổ"

    Thứ tự ưu tiên phân loại (từ cao xuống thấp):
    1. Có "sổ hồng" hoặc "sổ đỏ"
       → "Sổ đỏ / Sổ hồng"

    2. Không có sổ, nhưng có "hợp đồng mua bán"
       → "Hợp đồng mua bán"

    3. Không có sổ, không có HĐMB, nhưng có "giấy viết tay"
       → "Giấy viết tay"

    4. Không có các trường hợp trên, nhưng có "đang chờ sổ"
       → "Đang chờ sổ"

    5. Không xác định được
       → NaN

    Parameters
    ----------
    text : str hoặc None
        Chuỗi mô tả pháp lý của bất động sản
        (ví dụ: "Nhà có sổ hồng riêng", "Hợp đồng mua bán, đang chờ sổ").

    Returns
    -------
    str hoặc numpy.nan
        Nhãn pháp lý đã được chuẩn hóa, hoặc NaN nếu không xác định được.

    Notes
    -----
    - Văn bản được chuyển về chữ thường (lowercase) trước khi xử lý.
    - Hàm chỉ dựa trên từ khóa, không phân tích ngữ nghĩa sâu.
    - Trường hợp văn bản chứa nhiều từ khóa,
      kết quả phụ thuộc vào **thứ tự ưu tiên** đã định nghĩa.
    - Phù hợp dùng trong bước tiền xử lý dữ liệu (feature engineering)
      cho các bài toán phân tích bất động sản.
    """

    if text is None:
        return np.nan

    t = text.lower()

    has_so_hong = "sổ hồng" in t
    has_so_do   = "sổ đỏ" in t
    has_hdmua   = "hợp đồng mua bán" in t
    has_giayvt  = "giấy viết tay" in t
    has_dangcho = "đang chờ sổ" in t

    # Ưu tiên: nếu có sổ hồng / sổ đỏ → gán Sổ đỏ / Sổ hồng
    if has_so_hong or has_so_do:
        return "Sổ đỏ / Sổ hồng"

    # Nếu không có sổ mà có hợp đồng mua bán
    if has_hdmua:
        return "Hợp đồng mua bán"

    # Nếu không có sổ, không có HĐMB mà có giấy viết tay
    if has_giayvt:
        return "Giấy viết tay"

    if has_dangcho:
        return "Đang chờ sổ"

    # Không có gì → giữ NaN
    return np.nan
```

## n. Chuẩn hoá thuộc tính "total_bedroom" và "total_restroom"

```python
def extract_number(text):
    """
    Trích xuất số nguyên đầu tiên xuất hiện trong chuỗi văn bản.

    Hàm này tìm và lấy **chuỗi chữ số liên tiếp đầu tiên** trong `text`,
    sau đó chuyển sang kiểu số nguyên (`int`).
    Thường được dùng để chuẩn hóa các cột dạng mô tả
    như số phòng ngủ, số nhà vệ sinh.

    Parameters
    ----------
    text : str
        Chuỗi văn bản chứa số
        (ví dụ: "3 phòng ngủ", "WC: 2", "1 toilet").

    Returns
    -------
    int hoặc None
        - Giá trị số nguyên đã trích xuất.
        - Trả về None nếu không tìm thấy số trong chuỗi.

    Notes
    -----
    - Chỉ trích xuất **số đầu tiên** xuất hiện trong chuỗi.
    - Không xử lý số thập phân (ví dụ: "1.5").
    - Không xử lý các trường hợp phức tạp như:
        + "3–4 phòng"
        + "2 phòng + 1 gác"
    - Phù hợp dùng trong bước tiền xử lý dữ liệu (feature engineering)
      cho các cột như `total_bedroom`, `total_restroom`.
    """

    match = re.search(r'\d+', str(text))
    return int(match.group()) if match else None

df['total_bedroom'] = df['total_bedroom'].apply(extract_number)
df['total_restroom'] = df['total_restroom'].apply(extract_number)
```
