```python
response = requests.get ('https://thuviennhadat.vn/ban-nha-dat-toan-quoc?trang=1')
soup = BeautifulSoup (response.text, "html.parser")
base_url = "https://thuviennhadat.vn"
# b = soup.select ("aside[class='ui segment mobile-container-list'] > a")
# for c in b:
    
#     print (c.get('href'))
# print(soup.prettify())
articles = soup.select ("div[class='navigateTo mb-mt-15'] > div[class='ui divided items mobile-not-show']")
# articles = soup.select('aside[class="ui segment mobile-container-list"] > a')
print (len (articles))
for article in articles:
    # Get URL
    link = article.select_one ('a').get ("href")
    if link and link.startswith("/"):
        link = base_url + link

    # Get Post Date
    post_date = article.select_one (
        "div.KinhDoanhNhaDat-index-post-footer span"
    )
    post_date = post_date.get_text(strip=True) if post_date else None
    
    # Get Location
    location = article.select_one ("div.ui.labeled.icon span")
    location = location.get_text(strip=True) if location else None
    
    print (post_date)
```

```python
from pymongo import MongoClient
import pandas as pd

client = MongoClient("mongodb://localhost:27017/")
db = client["Finance_News"]

collections = ["CafeF"]

dfs = []
for col_name in collections:
    col = db[col_name]
    data = list(col.find())  # list các dict
    df = pd.json_normalize(data)  # chuyển dict lồng dict thành dataframe
    dfs.append(df)

# Ghép các dataframe lại
full_df = pd.concat(dfs, ignore_index=True)
full_df
```
